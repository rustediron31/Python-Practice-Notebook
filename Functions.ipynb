Define a function that accepts a number and returns whether the number is even or odd.

```python
def num(number):
  if number % 2 == 0:
    return "even"
  else:
    return "odd"
```

```python
num(11)
```

Define a function to create and print a list where the values are square of numbers between 1 and 30 (both included).

```python
def square_values():
  square = [x**2 for x in range(1,30)]
  print(square)



```

```python
def by_three():
  number = int(input("Enter the number = "))

  if number%3==0:
    print("The number is divisible of 3", "The sqaure root of the number", number**3)

  else:
    print("The number is not divisible of 3", "The sqaure root of the number", number**3)

```

```python
by_three()
```
