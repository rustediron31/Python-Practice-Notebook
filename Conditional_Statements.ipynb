Take an alphabet as input from the user and check if the alphabet is a vowel, consonant or an exception


Vowel - 'a','e','i','o','u'

Exception - 'h', 'y'

Consonant - rest other alphabets

```python

alphabet = str(input("Enter the alphabet you choose = "))

if alphabet in ('a','e','i','o','u'):
  print("Vowel")

elif alphabet in ('h', 'y'):
  print("Exception")

else:
  print("Consonant")
```

Write a program to check whether a number given by a user is divisible by 2 and 3 both

```python
number = int(input("Enter the number = "))

if number%2 == 0 and number%3 == 0:
  print (number, "is divisable of 2 and 3")

else:
  print (number, "is not divisable of 2 and 3")
```

Write a program to accept the cost price  of a bike as input and display the road tax to be paid according to the following crieteria :-

Cost price - 100000 Tax - 15%

Cost price - between 50000 to 100000 Tax - 10%

Cost price - less than 50000 Tax - 5%

```python
cost_of_bikes = float(input("Enter the cost of the bike: "))

if cost_of_bikes > 100000:
    tax = cost_of_bikes * 0.15
elif 50000 < cost_of_bikes <= 100000:
    tax = cost_of_bikes * 0.10
else:
    tax = cost_of_bikes * 0.05

total_cost = cost_of_bikes + tax
print(f"Total tax = ₹{tax}, Total cost of Bike with tax = ₹{total_cost}")

```

Write a program to calculate the electricity bill by accepting the number of units consumed by the user as input. Use the following price chart for reference :-

Price of first 100 units - no charge

Price of next 100 units - Rs 5/- per unit

After 200 units - Rs 10/- per unit

```python
units_consumed = float(input("Enter the units consumed = "))

if units_consumed <= 100:
  charges = 0 #no charges under 100 units

elif 100 < units_consumed <= 200: # 5rs per unit for units between 100 to 200
  charges = units_consumed * 5

else:
  charges = units_consumed * 10 # 10rs per unit for units between more than 200


print(f"total charges: ${ charges:.2f}")
```
